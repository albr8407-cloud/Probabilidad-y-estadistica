```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
# Carga el archivo CSV en una variable llamada covid
covid = pd.read_csv("20-06-30-COVID19MEXICO - codificado.csv", encoding = "ISO-8859-1")
```

```python
# Mostramos los primeros 5 registros del archivo
covid.head()
```

```python
# Mostrar la informacion la informacion del archivo
# Muestra el numero de registros, tipo de datos de la columna, valores nulos etc.
covid.info()
```

```python
# Mostrar la edad maxima de la base de datos de la columna edad
covid.EDAD.max()
```

```python
# Mostrar la edad minima de la base de datos de la columna edad
covid.EDAD.min()
```

```python
# Dibuja el histograma del atreibuto "EDAD" con bins = 3
# Calcular la media y guardarla en una variable para mostrarla en el histograma
media = covid.EDAD.mean()
plt.pyplot.hist(covid.EDAD, bins = 11)
plt.pyplot.title("Histograma edad pruebas COVID")
plt.pyplot.axvline(media, color = 'y', linestyle = 'dashed', linewidth=2, label=f'Media: {media:.2f}')
plt.pyplot.xlabel("Edad")
plt.pyplot.ylabel("Frecuencia")
```

```python
# Dibuja el histograma de la columna sexo,
#esto permite comparar cuantos hay en cada grupo
plt.pyplot.hist(covid.SEXO, bins = 3)

# establece las etiquetas x/y y muestra el titulo
plt.pyplot.title("Histograma Sexo de Pruebas COVID")
plt.pyplot.xlabel("Sexo")
plt.pyplot.ylabel("Frecuencia")
```

```python
# Dibuje el histograma del atributo "Tipo Paciente, esto permite
# comparar cuantos hay en cada grupo
plt.pyplot.hist(covid.TIPO_PACIENTE, bins = 3)

# establece las etiquetas x/y y muestra el titulo
plt.pyplot.title("Histograma Tipo Paciente de Pruebas COVID")
plt.pyplot.xlabel("Tipo Paciente")
plt.pyplot.ylabel("Frecuencia")
```

```python
# Crear un diagrama de cajas para visualizar la distribucion
# de los rasgos de edad por sexo
sns.boxplot(x ="SEXO", y ="EDAD", data=covid)
```

```python
# Crear un diagrama de cajas para visualizar la distribucion
# de los rasgos de la edad por sexo
sns.boxplot(x="TIPO_PACIENTE", y ="EDAD", data=covid)
```

```python
# Crear un diagrama de cajas para visualizar la distribucion de
# los rasgos de edad por tipo de paciente

sns.boxplot(x="RESULTADO", y ="EDAD", data=covid)
```
